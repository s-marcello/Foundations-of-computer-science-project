### 1. Normalize the location field of *Users* dataset, splitting into city, region, country.

Creo un'espressione regolare capace di discriminare tra stringhe che contengono al massimo tre parole divise da una o più virgole o uno o più spazi

```python
compile1 = re.compile(r'^(?P<City>[^,]*),+\s*(?P<Region>[^,]*),+\s*(?P<Country>[^,]*)$')
```

Controllo la presenza di maiscuole

```python
Users[Users['Location'].str.contains(r'[A-Z]+')]
```

Divido il dataset sulla base del contenuto della stringa relativa alla colonna 'location'


```python
location3fields = Users[Users['Location'].str.contains(compile1) == True]
locationmessy = Users[Users['Location'].str.contains(compile1) == False]

```

```python
len(location3fields) / len(Users) * 100

```

```python
location3fields[['City','Region','Country']] = location3fields['Location'].str.extract(compile1)
location3fields

```

### 2. For each book in the *Books* dataset, compute its average rating.

In primis unisco i dataset Books e Books_ratings, mantenendo tutti i libri contenuti in Books.

```python
Books_and_ratings = pd.merge(Books, Books_ratings, on='ISBN', how= 'left')
```

Raggruppo, trovo la media e arrotondo

```python
Books_and_average = Books_and_ratings.groupby(['ISBN','Book-Title','Book-Author'], as_index=False) \
[['ISBN','Book-Title','Book-Author','Book-Rating']].mean()
Books_and_average.rename({'ISBN':'ISBN','Book-Title':'Book-Title','Book-Author':'Book-Author','Book-Rating':'Book-Average-Rating'},
                         axis='columns', inplace= True)
Books_and_average['Book-Average-Rating'] = round(Books_and_average['Book-Average-Rating'], ndigits= 1)
Books_and_average.head(15)
```

### 3. For each book in the *GoodBooks* dataset, compute its average rating.

Il campo work_rating_count sembra essere una somma dei campi ratings(1,2,3,4,5). 


Immagino che il suffisso numerico stia a indicare il voto della recensione.


Creo una nuova colonna average.

```python

Goodbooks['average'] = ((Goodbooks['ratings_1'] * 1 )+(Goodbooks['ratings_2'] * 2 )+
                        (Goodbooks['ratings_3'] * 3 )+(Goodbooks['ratings_4'] * 4 )+
                        (Goodbooks['ratings_5'] * 5 ))/Goodbooks['work_ratings_count']
Goodbooks['average'] = round(Goodbooks['average'], ndigits= 1)
```

Raggruppo e faccio la media

```python
Goodbooks_and_average = Goodbooks.groupby(['book_id','isbn','original_title','title'],
                                          as_index= False)[['book_id','isbn','original_title','title','average']].mean()
Goodbooks_and_average.head(20)
```

### 4. Merge together all rows sharing the same book title, author and publisher. We will call the resulting datset `merged books`. The books that have not been merged together will not appear in `merged books`.

Raggruppo, poi considero solo i libri con più di un'edizione, escludendo gli altri dal dataset.

```python
Groupby_books = Books.groupby(['Book-Title','Book-Author','Publisher'], as_index= False)['ISBN'].count()
merged_books = Groupby_books[Groupby_books['ISBN'] > 1]
merged_books.drop(['ISBN'], axis= 'columns', inplace= True)
```

```python
merged_books.head(20)
```

### 5. For each book in `merged books` compute its average rating.

The average is computed considering all books in `books` that have been merged.

Tutti i libri contenuti in merged_books hanno più di un'edizione, il voto di queste varie edizioni è contenuto nel dataset Books_Ratings.


Per prima cosa, per ogni libro contenuto in merged_books, bisogna recuperare tutte le edizioni dal dataset Books.

```python
merged_books_and_info = pd.merge(merged_books, Books, on= ['Book-Title','Book-Author','Publisher'], how='left')
merged_books_and_info.head(20)
```

```python
merged_books_and_info.drop(['Image-URL-S', 'Image-URL-M', 'Image-URL-L'], axis= 1, inplace= True)
```

Ad ogni libro ed ogni edizione associamo il voto contenuto nel dataset Books_ratings.

```python
merged_books_and_ratings = pd.merge(merged_books_and_info, Books_ratings, on='ISBN', how='left')
merged_books_and_average = merged_books_and_ratings.groupby(
    ['Book-Title','Book-Author','Publisher','ISBN'], as_index=False) \
[['Book-Title','Book-Author','Publisher','ISBN','Book-Rating']].mean()
merged_books_and_average.rename(
    {'Book-Title':'Book-Title','Book-Author':'Book-Author','Publisher':'Publisher','ISBN':'ISBN','Book-Rating': 'Book-Average'},
    axis= 1, inplace= True)
merged_books_and_average.head(20)
```

### 6. For each book in `merged books` compute the minimum and maximum of the average ratings over all corresponding books in the `books` dataset.

Hence for each book in `merged books` we will have exactly two values (a minimum and a maximum)

È possibile compiere contemporaneamente più operazioni su un raggruppamento, a noi interessano il minimo ed il massimo.

```python
merged_books_and_average_agg = merged_books_and_average.groupby(['Book-Title','Book-Author','Publisher']).agg(
    {'Book-Average':[np.min, np.max, np.mean, 'count']})
merged_books_and_average_agg.head(15)
```

### 7. For each book in `goodbooks`, compute the list of its authors. Assuming that the number of reviews with a text (column `work_text_reviews_count`) is split equally among all authors, find for each authors the total number of reviews with a text. We will call this quantity the *shared number of reviews with a text*.

Creo una copia del dataset Goodbooks, creo tre nuove variabili: la lista degli autori, il numero d'autori per libro ed il numero di recensioni diviso il numero di autori.

```python
Goodbooks_slim = Goodbooks[['authors','original_publication_year','title','average','work_text_reviews_count']]
Goodbooks_slim['authors'] = Goodbooks_slim['authors'].str.replace(r'^\s', '', regex= True)
```

```python
Goodbooks_slim['list_authors'] = Goodbooks_slim['authors'].str.split(r',\s')
Goodbooks_slim['number_authors'] = Goodbooks_slim['list_authors'].apply(lambda x: len(x))
Goodbooks_slim['text_reviews_per_author'] = round(Goodbooks_slim['work_text_reviews_count'] / Goodbooks_slim['number_authors'],
                                                  ndigits= 2)
```

Estraggo tutti gli autori

```python
autore = []
for riga in Goodbooks_slim['list_authors']:
    for elemento in riga:
        if elemento not in autore:
            autore.append(elemento)
autore.sort()
autore
```

Creo una funzione che assegna ad ogni autore il suo numero di recensioni e costruisco un nuovo dataset contenente gli autori ed il numero di recensioni associate.

```python
shared_review= np.zeros(109)
def shared_number_func(x):
    for i in range(len(autore) -1):
        if autore[i] in x['list_authors']:
            shared_review[i] = shared_review[i] + x['text_reviews_per_author']
```

```python
Goodbooks_slim.apply(shared_number_func, axis= 1) ;
```

```python
shared_review
shared_number_of_reviews_with_a_text = list(shared_review)
```

```python
len(shared_number_of_reviews_with_a_text)
```

```python
id_author = list(range(109))
len(id_author)
```

```python
Authors = pd.DataFrame(data={'id_author' : id_author,'author' : autore,
                             'shared_number_of_reviews_with_a_text' : shared_number_of_reviews_with_a_text })
Authors.head(10)
```

### 8. For each year of publication, determine the author that has the largest value of the shared number of reviews with a text.

```python
Goodbooks_year_sort = Goodbooks_slim.sort_values('original_publication_year')
Goodbooks_year_sort
```

Creo una funzione che trasferisce in una matrice anni per autori il numero di recensioni attribuibili ad ogni autore, poi per ogni anno stabilisco quale sia l'autore con più recensioni condivise, trovando il massimo.

Creo un dataset con i risultati ottenuti.

```python
max_shared_review = np.zeros([58, 109])
anni = list(Goodbooks_year_sort['original_publication_year'].unique())

def func_max_per_year(x): 
    for an in range(len(anni)):
        if x['original_publication_year'] == anni[an]: 
            for au in range(len(autore)):
                if autore[au] in x['list_authors']:
                      if x['text_reviews_per_author'] > max_shared_review[an,au]:
                        max_shared_review[an,au] = x['text_reviews_per_author']         
```

```python
Goodbooks_year_sort.apply(func_max_per_year, axis= 1) ;
```

```python
author_of_the_year = []
shared_number_of_reviews = []

for i in range(len(max_shared_review)):
    best_author_year = max_shared_review[i].argmax()
    author_of_the_year.append(autore[best_author_year])
    number_of_review = max_shared_review[i].max()
    shared_number_of_reviews.append(number_of_review)       
    
authors_of_the_year = pd.DataFrame(data= {'year': anni, 'author': author_of_the_year,
                                          'shared_number_of_reviews': shared_number_of_reviews})
authors_of_the_year.tail(10)
```
